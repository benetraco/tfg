```python
import matplotlib.pyplot as plt
import pandas as pd

# --- Define the data ---

# Guidance values
guidance = [0, 1, 2, 3, 5, 7, 10]

# FID scores for each scanner (VH, SHIFTS, WMH)
FID = {
    "VH": [48.56, 32.49, 36.29, 37.29, 45.08, 50.10, 53.49],
    "SHIFTS": [40.89, 34.16, 38.30, 49.82, 50.51, 50.07, 50.53],
    "WMH": [48.48, 34.72, 36.01, 38.21, 42.51, 47.57, 55.96]
}

# CMMD scores for each scanner (VH, SHIFTS, WMH)
CMMD = {
    "VH": [0.382, 0.342, 0.365, 0.317, 0.343, 0.384, 0.404],
    "SHIFTS": [0.325, 0.328, 0.325, 0.317, 0.366, 0.358, 0.400],
    "WMH": [0.502, 0.402, 0.393, 0.444, 0.430, 0.504, 0.519]
}

# LPIPS scores for each scanner (VH, SHIFTS, WMH)
LPIPS = {
    "VH": [0.226, 0.194, 0.192, 0.193, 0.202, 0.210, 0.224],
    "SHIFTS": [0.251, 0.245, 0.253, 0.258, 0.256, 0.258, 0.261],
    "WMH": [0.248, 0.232, 0.234, 0.242, 0.255, 0.271, 0.280]
}

# Common larger font sizes
title_fontsize = 20
label_fontsize = 16
tick_fontsize = 14
legend_fontsize = 14

# --- Plot CMMD ---
plt.figure(figsize=(10, 5))
for scanner in CMMD:
    plt.plot([f"g{g}.0" for g in guidance], CMMD[scanner], marker='o', label=scanner)
plt.title("CMMD Score Across Guidance Values", fontsize=title_fontsize)
plt.xlabel("Guidance Value", fontsize=label_fontsize)
plt.ylabel("CMMD", fontsize=label_fontsize)
plt.xticks(fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)
plt.legend(fontsize=legend_fontsize)
plt.grid(True)
plt.show()

# --- Plot FID ---
plt.figure(figsize=(10, 5))
for scanner in FID:
    plt.plot([f"g{g}.0" for g in guidance], FID[scanner], marker='o', label=scanner)
plt.title("FID Score Across Guidance Values", fontsize=title_fontsize)
plt.xlabel("Guidance Value", fontsize=label_fontsize)
plt.ylabel("FID", fontsize=label_fontsize)
plt.xticks(fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)
plt.legend(fontsize=legend_fontsize)
plt.grid(True)
plt.show()

# --- Plot LPIPS ---
plt.figure(figsize=(10, 5))
for scanner in LPIPS:
    plt.plot([f"g{g}.0" for g in guidance], LPIPS[scanner], marker='o', label=scanner)
plt.title("LPIPS Score Across Guidance Values", fontsize=title_fontsize)
plt.xlabel("Guidance Value", fontsize=label_fontsize)
plt.ylabel("LPIPS", fontsize=label_fontsize)
plt.xticks(fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)
plt.legend(fontsize=legend_fontsize)
plt.grid(True)
plt.show()
```
